```python
import os
import sympy as smp
import numpy as np
from scipy.special import *
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "text.latex.preamble": r"\usepackage{amsfonts}"
})

# Get absolute path to current directory
current_directory = os.getcwd()
print(current_directory)
```

# Laguerre Polynomials

```python
def get_laguerre_polynomial(n: int):
    x = smp.symbols('x')
    L = smp.symbols(f'L_{n}', cls=smp.Function)
    L = L(x)
    eq = smp.Eq(L, smp.laguerre(n, x))
    display(eq)


get_laguerre_polynomial(7)
```

```python
def plot_laguerre_polynomials(l : list):
    x = np.linspace(0, 4, 1000)

    n_values = l

    plt.figure()

    for n in n_values:
        plt.plot(x, laguerre(n)(x), label=f'$n={n}$')

    plt.xlabel(r'$x$')
    plt.ylabel(r'$L_{n}(x)$')
    plt.title(r'Laguerre Polynomials $L_{n}(x), n \in \mathbb{N}$')
    plt.legend(loc='best')

    plt.grid(True, color='lightgray')
    plt.savefig('images/laguerre-polynomials.pdf', format='pdf')

    plt.show()

plot_laguerre_polynomials(range(6))
```

# Associated Laguerre Polynomials

```python
def get_laguerre_polynomial(n: int, k = smp.Symbol('k')):
    x = smp.symbols('x')
    L = smp.symbols(f'L_{n}^{k}', cls=smp.Function)
    L = L(x)
    eq = smp.Eq(L, smp.assoc_laguerre(n, k, x))
    display(eq)

# args: n, k (optional)
get_laguerre_polynomial(2, 5)
```

```python
def plot_laguerre_polynomials(k_value, l : list):
    x = np.linspace(0, 4, 1000)

    n_values = l

    plt.figure()

    for n in n_values:
        plt.plot(x, assoc_laguerre(x, n, k_value), label=f'$n={n}$')

    plt.xlabel(r'$x$')
    plt.ylabel(f'$L_n^{k_value}(x)$')
    plt.title(r'Associated Laguerre Polynomials $L_{n}^{k}(x) = L_{n}^{(\alpha)},$ for which $n, k \in \mathbb{N}$')
    plt.legend(loc='best')

    plt.grid(True, color='lightgray')
    plt.savefig(f'images/assoc-laguerre-polynomials-{k_value}.pdf', format='pdf')

    plt.show()

plot_laguerre_polynomials(2, range(6))
```

# Laguerre's Associated Equation Main Family

The most common famlily of the Laguerre's Associated Equation is defined as 

$$\psi_p^k (x) = x^{(k+1)/2} e^{-x/2} L_p^k$$

such that the function $\psi_p^k (x)$ satisfies:

$$\ddot{\psi}_p^k (x) + \psi_p^k(x) \left[\frac{-(k^2-1)}{4x^2} + \frac{2p+k+1}{2x} - \frac{1}{4}\right] = 0$$
