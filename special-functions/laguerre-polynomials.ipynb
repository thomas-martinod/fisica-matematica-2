```python
import sympy as smp
import numpy as np
from scipy.special import *
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "text.latex.preamble": r"\usepackage{amsfonts}"
})
```

# Laguerre Polynomials

```python
def get_laguerre_polynomial(n: int):
    x = smp.symbols('x')
    L = smp.symbols(f'L_{n}', cls=smp.Function)
    L = L(x)
    eq = smp.Eq(L, smp.laguerre(n, x))
    display(eq)


get_laguerre_polynomial(7)
```

```python
def plot_laguerre_polynomials(l : list):
    x = np.linspace(0, 4, 1000)

    n_values = l

    plt.figure()

    for n in n_values:
        plt.plot(x, laguerre(n)(x), label=f'$n={n}$')

    plt.xlabel(r'$x$')
    plt.ylabel(r'$L_{n}(x)$')
    plt.title(r'Laguerre Polynomials $L_{n}(x), n \in \mathbb{N}$')
    plt.legend(loc='best')

    plt.grid(True, color='lightgray')
    plt.savefig('special-functions/laguerre-polynomials.pdf', format='pdf')
    plt.show()


plot_laguerre_polynomials(range(6))
```
